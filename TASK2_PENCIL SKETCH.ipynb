### Importing required libraries

```python
import cv2
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display
```

### Read Image and Display

```python
img = cv2.imread('image1.jpeg',1)
img.shape
```

```python
cv2.imshow('Image',img)  # opens a pop up 
cv2.waitKey(0)  # This will close one particular image opened
cv2.destroyAllWindows()  # Closes all the image pop ups opened all at time
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img1))   # Displaying the image
```

- orginal image in RGB Format

### Converting RGB image into Grayscale image

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("New img", gray_image)
cv2.waitKey(0)
```

```python
display(Image.fromarray(gray_image))
```

- Image in grayscale

```python
# invert the grayscale image
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
```

```python
display(Image.fromarray(inverted_image))
```

- Negative of an image

```python
# blur the inverted image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

### Converting blurred image to a Pencil sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
display(Image.fromarray(inverted_blurred))
```

- Blurred image

```python
display(Image.fromarray(pencil_sketch))
```

- Converted pencil sketch image

### Comparing original image with the pencil sketch

```python
cv2.imshow("original image", img)
cv2.imshow("pencil sketch", pencil_sketch)
cv2.waitKey(0)
```

```python
display(Image.fromarray(img))
display(Image.fromarray(pencil_sketch))
```

```python
res = np.hstack((gray_image,pencil_sketch))
cv2.imshow('im1',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
display(Image.fromarray(res))
```

- Conclusion : 
    - Initially, converted RGB Image to a grayscale image.
    - next, converted grayscale image to an inverted image to obtain negative of an image.
    - next, converted inverted image to a blur image.
    - Finally, converted the blur image to the required pencil sketch
